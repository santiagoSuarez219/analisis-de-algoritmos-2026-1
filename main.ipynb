```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import random
import time
```

```python
cantidad_de_datos = [2**i for i in range(1,15)] # [2, 4, 8, ..., ]
print(cantidad_de_datos)
```

```python
def calcular_complejidad_temporal(fn, cantidad_de_datos, desordenar=False, invertir=False):
    tiempo = []
    for n in cantidad_de_datos: 
        arr = [i for i in range(n)]
        if desordenar:
            random.shuffle(arr)
        if invertir:
            arr.reverse()
        tiempo_inicial = time.time()
        fn(arr)
        tiempo_final = time.time()
        tiempo_total = tiempo_final - tiempo_inicial
        tiempo.append(tiempo_total)
    return tiempo
    
```

```python
# Algoritmo de ordenamiento por inserción
def insertion_sort(arr):
    for j in range(1, len(arr)):
        key = arr[j]
        i = j - 1
        while i >= 0 and arr[i] > key:
            arr[i + 1] = arr[i]
            i -= 1
        arr[i + 1] = key
    return arr
```

```python
tiempo_omega = calcular_complejidad_temporal(insertion_sort, cantidad_de_datos)
tiempo_big_o = calcular_complejidad_temporal(insertion_sort, cantidad_de_datos, invertir=True)
sns.lineplot(x=cantidad_de_datos, y=tiempo_omega, label="Complejidad Omega")
sns.lineplot(x=cantidad_de_datos, y=tiempo_big_o, label="Complejidad BiG O")
plt.xlabel("Cantidad de datos (N)")
plt.ylabel("Complejidad temporal (s)")
plt.title("Analisis de notacion Omega")
plt.grid()
plt.show()
```
