# Notacion asintotica

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import random
import time
```

## Algoritmo de ordenamiento por inserciÃ³n

```python
def insertion_sort(arr):
    for j in range(1, len(arr)):
        key = arr[j]
        i = j - 1
        while i >= 0 and arr[i] > key:
            arr[i + 1] = arr[i]
            i -= 1
        arr[i + 1] = key
    return arr
```

```python
cantidad_de_datos = [2**i for i in range(1, 15)]
```

```python
def calcular_complejidad_temporal(fn, shuffle=False, invert = False, cantidad_de_datos=cantidad_de_datos):
    tiempo = []
    for n in cantidad_de_datos:
        arr = [i for i in range(n)]
        if shuffle:
            random.shuffle(arr)
        if invert:
            arr.reverse()
        start_time = time.time()
        fn(arr)
        end_time = time.time()
        tiempo.append(end_time - start_time)
    return tiempo
```

```python
tiempos = []
for i in range(3):
    tiempo = calcular_complejidad_temporal(insertion_sort, shuffle=True)
    tiempos.append(tiempo)
```

```python
sns.lineplot(x=cantidad_de_datos, y=tiempos[0], label='EjecuciÃ³n 1')
sns.lineplot(x=cantidad_de_datos, y=tiempos[1], label='EjecuciÃ³n 2')
sns.lineplot(x=cantidad_de_datos, y=tiempos[2], label='EjecuciÃ³n 3')
plt.xlabel('Cantidad de datos')
plt.ylabel('Tiempo (segundos)')
plt.title('Tiempo de ejecuciÃ³n vs Cantidad de datos')
plt.grid()
plt.show()
```

La notaciÃ³n asintÃ³tica es el lenguaje matemÃ¡tico que usamos para describir quÃ© tan rÃ¡pido crece el costo de un algoritmo cuando el tamaÃ±o del problema aumenta considerablemente.

$$T(n)âÎ(g(n))$$

## notacion Big O
La **notaciÃ³n Big O** describe **quÃ© tan rÃ¡pido crece el tiempo de ejecuciÃ³n de un algoritmo en el peor escenario**, cuando el tamaÃ±o de la entrada aumenta.

En tÃ©rminos simples:

> Big O nos dice el lÃ­mite superior del crecimiento de un algoritmo cuando el problema se vuelve grande.
> 

$$
c_1 * g(n) \ge T(n)
$$

Supongamos que se tiene un algoritmo con una complejidad temporal que puede ser reprentado por la funcion

$$T\left(n\right) = 3n^{2} + 2n + 100$$

```python
t_n = [3*n**2 + 2*n + 100 for n in cantidad_de_datos]
sns.lineplot(x=cantidad_de_datos, y=t_n, label='T(n)', color='blue', linestyle='--', marker='o')
plt.xlabel('Cantidad de datos')
plt.ylabel('T(n)')
plt.title('T(n) vs Cantidad de datos')
plt.grid()
plt.show()
```

```python
t_n = [3*n**2 + 2*n + 100 for n in cantidad_de_datos]
constantes_c = [i for i in range(10)]

sns.lineplot(x=cantidad_de_datos, y=t_n, label='T(n)', color='blue', linestyle='--', marker='o')
for constante_c in constantes_c:
    t_n_constante = [constante_c*n**2 for n in cantidad_de_datos]
    sns.lineplot(x=cantidad_de_datos, y=t_n_constante, label=f'T(n) con c={constante_c}', linestyle='-')
plt.xlabel('Cantidad de datos')
plt.ylabel('T(n)')
plt.title('T(n) vs Cantidad de datos con diferentes constantes c')
plt.grid()
plt.legend()
plt.show()
```

```python
t_n = [3*n**2 + 2*n + 100 for n in cantidad_de_datos]
t_n_2 = [3*n**2 + 2*n for n in cantidad_de_datos]
sns.lineplot(x=cantidad_de_datos, y=t_n, label='T(n)', color='blue', linestyle='--', marker='o')
sns.lineplot(x=cantidad_de_datos, y=t_n_2, label='3n^2 + 2n', color='red', linestyle='-')
plt.xlabel('Cantidad de datos')
plt.ylabel('T(n)')
plt.title('T(n) vs Cantidad de datos')
plt.grid()
plt.show()
```

En notacion asintoctica, solo nos interesa el termino que mas crece, por lo tanto, podemos decir que:
$$T(n)âO(n^2)$$

Big O solo dice:

Piensa en Big O como:

- Una **barrera de velocidad mÃ¡xima**
- Un **techo de crecimiento**

## NotaciÃ³n Î© (Omega)
La **notaciÃ³n Omega** describe **quÃ© tan rÃ¡pido crece el tiempo de ejecuciÃ³n de un algoritmo en el mejor escenario**, cuando el tamaÃ±o de la entrada aumenta.

En tÃ©rminos simples:

> Omega nos dice el lÃ­mite inferior del crecimiento de un algoritmo cuando el problema se vuelve grande.

$$
c_2 * g(n) \le T(n)
$$

## Calculando manualmente la notacion asintotica

```python
def f(n):
    for i in range(len(n)): # n
        print(i) # n
    for j in range(len(n)): # n
        print(j) # n

# n + n + n + n = 4n = T(n) -> O(n)
```

```python
cantidad_de_datos = [2**i for i in range(1, 15)]
```

```python
tiempos = calcular_complejidad_temporal(f, shuffle=False)
sns.lineplot(x=cantidad_de_datos, y=tiempos, label='T(n)', color='blue', marker='o')
plt.xlabel('Cantidad de datos')
plt.ylabel('Tiempo (segundos)')
plt.title('Tiempo de ejecuciÃ³n de la funciÃ³n f(n)')
plt.grid()
plt.show()
```

```python
def g(n):
    for i in range(len(n)): # n
        for j in range(i): # n * n = n^2
            print(i, j) # n^2

# n + n*n = n^2 + n = T(n) -> O(n^2)
```

```python
cantidad_de_datos = [2**i for i in range(1, 9)]
```

```python
cantidad_de_datos
```

```python
tiempos = calcular_complejidad_temporal(g, shuffle=False, cantidad_de_datos=cantidad_de_datos)
sns.lineplot(x=cantidad_de_datos, y=tiempos, label='T(n)', color='blue', marker='o')
plt.xlabel('Cantidad de datos')
plt.ylabel('Tiempo (segundos)')
plt.title('Tiempo de ejecuciÃ³n de la funciÃ³n f(n)')
plt.grid()
plt.show()
```

```python
def insertion_sort(arr):
    for j in range(1, len(arr)): # n
        key = arr[j] # n 
        i = j - 1 # n 
        while i >= 0 and arr[i] > key: # n * n = n^2
            arr[i + 1] = arr[i] # n^2
            i -= 1 # n^2
        arr[i + 1] = key # n
    return arr

# n + n + n*n + n*n + n = 2n^2 + 3n = T(n) -> O(n^2)
```
