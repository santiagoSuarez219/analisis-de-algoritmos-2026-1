**Variable Names and Programming Style**


- Classes: *UpperCamelCase* (e.g. ```class Car```, ```class MoneyCash```, ```class CartesianPoint```)
- Methods: *lowerCamelCase* (e.g. ```execute()```, ```runFast()```, ```setParameters()```, ```predict()```)
- Variables and functions: *lowercase* or *lowercase_separated_by_underscores* (e.g. ```flag```, ```temp_1```, ```speed```).  
Single-character variable names should be avoided unless they are temporary, in which case they are typically used. ```i```, ```j```, ```k```, ```m```, y ```n``` for integers, and ```c```, ```d```, y ```e``` for characters.
- Constants: *CAPITALIZED_WITH_UNDERSCORES* (e.g. ```MAX_TEMPERATURE```, ```N_JOBS```, ```LAYERS```)

## Operadores aritmeticos

````python
a = 10
b = 5

suma = a + b
resta = a - b
multiplicacion = a * b
division = a / b
potencia = a ** b
modulo = a % b
division_entera = a // b
````

## Operadores logicos

````python
a = 10
b = 5

a > b # Mayor que: True
a == b # Igual a: False
a != b # Diferente de: True

a > 5 and b < 10 # AND: True
a > 15 or b < 10 # OR: True
not(a > b) # NOT: False
````

### Assignment operators

- a += b (equivalent to a=a+b)
- a -= b (equivalent to a=a-b)
- a *= b (equivalent to a=a$\times $b)
- a /= b (equivalent to a=a/b)
- a //= b (equivalent to a=a//b)
- a **= b (equivalent to a=a$^b$)
- a %= b (equivalent to a=a%b)
